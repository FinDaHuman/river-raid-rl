# River Raid RL – Quickstart Notebook

This notebook walks through the core components of the project.
No environment or GPU required for cells 1–3.

## 1. Configuration

All hyper-parameters are stored in YAML files under configs/.

````python
import yaml
from pathlib import Path
ROOT = Path().resolve().parent
with open(ROOT / "configs" / "rainbow.yaml") as f:
    cfg = yaml.safe_load(f)
print("Agent:", cfg["agent"]["type"])
print("Hidden dim:", cfg["agent"]["params"]["hidden_dim"])
print("Total timesteps:", cfg["training"]["total_timesteps"])
````

## 2. Instantiate an Agent

Create a Rainbow agent with default config (CPU).

````python
import sys
sys.path.insert(0, str(ROOT))
from riverraid_rl.config import EnvConfig, RainbowConfig
from riverraid_rl.agents.rainbow import RainbowAgent

env_cfg = EnvConfig()
rb_cfg  = RainbowConfig()
agent   = RainbowAgent(env_cfg, rb_cfg, num_actions=6, device="cpu")
print("Agent created. Parameter count:", sum(p.numel() for p in agent.q_network.parameters()))
````

## 3. Try a Forward Pass on Dummy Data

The agent can process a batch of dummy observations without an environment.

````python
import numpy as np

dummy = np.zeros((4, 84, 84), dtype=np.uint8)
action = agent.act(dummy, training=False)
actions = ["NOOP", "FIRE", "UP", "RIGHT", "LEFT", "DOWN"]
print(f"Agent chose action {action} ({actions[action]})")
````

python experiments/train.py -c configs/rainbow.yaml -o runs/my_run
